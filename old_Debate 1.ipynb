```python
# function to get term after hashtag

def getHashtag(tweet):
    tweet = tweet.split()
    hashtags = []
    for item in tweet:
        if item[0] == "#":
            hashtags.append(item)
            
    return hashtags

def getRT(tweet):
    retweets = []    
    if tweet[0:1] == "RT":
        retweets.append(tweet)
    return retweets

def orgtweets(tweet):
    unique_tweets = []
    if tweet[0:1] != "RT":
        unique_tweets.append(tweet)
    return unique_tweets
        
# pd Series
hashtags = df.text.apply(getHashtag)
retweets = [element for element in df['text'].values if element.startswith('RT')]
orgtweets = [element for element in df['text'].values if not element.startswith('RT')]

# print(str(type(hashtags)))
# print(df.loc[2]['text'])
print(hashtags[:10])
print(str(len(orgtweets)))
print(str(len(retweets)))


    
```
